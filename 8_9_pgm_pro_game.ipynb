```python
import pygame
import pickle
pickled_model_5 = pickle.load(open('model_tesed_1_var_elii.pkl', 'rb'))
pickled_model = pickle.load(open('model_with_small_ver_auto.pkl', 'rb'))
note = {'Season': 'n',
 'MinTemp': 'n',
 'MaxTemp': 'n',
 'Sunshine':'n',
 'Rainfall': 'n',
 'Evaporation':'n',
 'WindSpeed':'n',
 'Humidity': 'n',
 'Pressure': 'n',
 'Cloud': 'n'}
note_state = note.copy()
kkk =  ['Season','MinTemp','MaxTemp', 'Sunshine','Rainfall', 'Evaporation',
       'WindSpeed', 'Humidity', 'Pressure',
       'Cloud']
def back():
    return {'Season': 'n',
 'MinTemp': 'n',
 'MaxTemp': 'n',
 'Sunshine':'n',
 'Rainfall': 'n',
 'Evaporation':'n',
 'WindSpeed':'n',
 'Humidity': 'n',
 'Pressure': 'n',
 'Cloud': 'n'}

def find_value_1(posterior_p):
    klo = str(posterior_p)
    aoo = []
    check = ['-','+','_','=','|',' ','\n']
    st = ''
    pos = 0
    for mk in klo:
        if(mk not in check):
            st += mk
            pos = 1
        elif(pos ==1): 
            #if(type(st)==type(1)):
            aoo.append(st) 
            st = ''
            pos = 0
    return aoo[5]

def find_output(note):
    for i in kkk:
        if(note[i]=='n'):
            del note[i]
    posterior_p_1 = pickled_model_5.query(['RainTomorrow'], evidence=note)
    print(posterior_p_1)
    print("\n")
    print("value of rain tomorrow: ")
    coco = find_value_1(posterior_p_1)
    print(coco)
    return coco

pygame.init()
screen = pygame.display.set_mode((1000, 700))
COLOR_INACTIVE = pygame.Color('lightskyblue3')
COLOR_ACTIVE = pygame.Color('dodgerblue2')
FONT = pygame.font.Font(None, 32)
buttons = pygame.sprite.Group()
class Button(pygame.sprite.Sprite):
    def __init__(self, screen,position, b_size,text, size):
        super().__init__()
        self.fg = COLOR_INACTIVE
        self.font = pygame.font.SysFont("Arial", size)
        self.text_render = self.font.render(text, 1, self.fg)
        self.image = self.text_render
        self.w,self.h = b_size
        self.x, self.y = position
        self.rect = pygame.Rect(self.x, self.y, self.w, self.h)
        self.position = position
        self.color = COLOR_INACTIVE
        buttons.add(self)
        self.update()
 
    def update(self):
        #self.fg, self.bg = self.colors.split(" on ")
        #screen.blit(self.text_render, (self.rect.x+5, self.rect.y+5))
        pygame.draw.rect(screen, self.color, self.rect, 2)
        #pygame.draw.rect(screen,self.bg, (self.x+10, self.y, self.w , self.h))
        
    def han_event(self,event):
        if event.type == pygame.MOUSEBUTTONDOWN:
            if self.rect.collidepoint(event.pos):
                #print(note)
                #print("clicked")
                #find_output(note)
                #note = back()
                #print(note)
        #return note
                return 1
        return 0

class InputBox:

    def __init__(self, x, y, w, h, name,text=''):
        self.rect = pygame.Rect(x, y, w, h)
        self.color = COLOR_INACTIVE
        self.text = text
        self.txt_surface = FONT.render(text, True, self.color)
        self.active = False
        self.name = name
        self.line2 = FONT.render("Season",True,COLOR_INACTIVE)
        self.line3 = FONT.render('MinTemp',True,COLOR_INACTIVE)
        self.line4 = FONT.render('MaxTemp',True,COLOR_INACTIVE)
        self.line5 = FONT.render('Sunshine',True,COLOR_INACTIVE)
        self.line6 = FONT.render('Rainfall',True,COLOR_INACTIVE)
        self.line7 = FONT.render('Evaporation',True,COLOR_INACTIVE)
        self.line8 = FONT.render('WindSpeed',True,COLOR_INACTIVE)
        self.line9 = FONT.render('Humidity',True,COLOR_INACTIVE)
        self.line10 = FONT.render('Pressure',True,COLOR_INACTIVE)
        self.line11 = FONT.render('Cloud',True,COLOR_INACTIVE)
        self.line12 = FONT.render('Rain Prob.',True,COLOR_INACTIVE)
        
        #self.txt_surface.blit()

    def handle_event(self, event):
        if event.type == pygame.MOUSEBUTTONDOWN:
            # If the user clicked on the input_box rect.
            if self.rect.collidepoint(event.pos):
                # Toggle the active variable.
                self.active = not self.active
            else:
                self.active = False
            # Change the current color of the input box.
            self.color = COLOR_ACTIVE if self.active else COLOR_INACTIVE
        if event.type == pygame.KEYDOWN:
            if self.active:
                #if event.key == pygame.K_RETURN:
                    #note[self.name] = int(self.text)
                    #print(self.text)
                    #self.text = ''
                if event.key == pygame.K_BACKSPACE:
                    self.text = self.text[:-1]
                else:
                    self.text += event.unicode
                # Re-render the text.
                self.txt_surface = FONT.render(self.text, True, self.color)
        

    def update(self):
        # Resize the box if the text is too long.
        width = max(200, self.txt_surface.get_width()+10)
        self.rect.w = width
        self.txt_surface = FONT.render(self.text, True, self.color)

    def draw(self, screen):
        # Blit the text.
        screen.blit(self.line2,(100,150))
        screen.blit(self.line3,(100,250))
        screen.blit(self.line4,(100,350))
        screen.blit(self.line5,(100,450))
        screen.blit(self.line6,(100,550))
        screen.blit(self.line7,(700,150))
        screen.blit(self.line8,(700,250))
        screen.blit(self.line9,(700,350))
        screen.blit(self.line10,(700,450))
        screen.blit(self.line11,(700,550))
        screen.blit(self.line12,(400,450))
        screen.blit(self.txt_surface, (self.rect.x+5, self.rect.y+5))
        # Blit the rect.
        pygame.draw.rect(screen, self.color, self.rect, 2)
        
    def get_val(self,note):
        if(self.text!=''):
            note[self.name] = int(self.text)
        self.text = ''
        
    def set_output(self,vall):
        self.text = str(vall)
        print("kfekfk e",vall)
    

def main(note):
    
    
    clock = pygame.time.Clock()
    input_box1 = InputBox(100, 200, 100, 32,'Season')
    input_box2 = InputBox(100, 300, 100, 32,'MinTemp')
    input_box3 = InputBox(100, 400, 100, 32,'MaxTemp')
    input_box4 = InputBox(100, 500, 100, 32,'Sunshine')
    input_box5 = InputBox(100, 600, 100, 32,'Rainfall')
    input_box6 = InputBox(700, 200, 100, 32,'Evaporation')
    input_box7 = InputBox(700, 300, 100, 32,'WindSpeed')
    input_box8 = InputBox(700, 400, 100, 32,'Humidity')
    input_box9 = InputBox(700, 500, 100, 32, 'Pressure')
    input_box10 = InputBox(700, 600, 100, 32,'Cloud')
    butt = Button(screen,(450,300),(80,40),"    get    ",28)
    input_box11 = InputBox(400, 500, 100, 32,'output')
    input_boxes = [input_box1, input_box2,input_box3,input_box4,input_box5,input_box6,input_box7,input_box8,input_box9,input_box10,input_box11]
    done = False

    while not done:
        for event in pygame.event.get():
            if event.type == pygame.QUIT:
                done = True
            for box in input_boxes:
                box.handle_event(event)
            if(butt.han_event(event)==1):
                for box in input_boxes[:-1]:
                    box.get_val(note)
                print(note)
                #print("clicked")
                input_box11.set_output(find_output(note))
                note = back()
                print(note)
                
        for box in input_boxes:
            box.update()

        screen.fill((30, 30, 30))
        for box in input_boxes:
            box.draw(screen)
            
            
            
        buttons.update()
        buttons.draw(screen)
        pygame.display.update()
        pygame.display.flip()
        clock.tick(30)
        


if __name__ == '__main__':
    main(note)
    pygame.quit()
```
